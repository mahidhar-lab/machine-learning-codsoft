```python
# ╔══════════════════════════════════════════════════════════════════════════════╗
# ║          CREDIT CARD FRAUD DETECTION — PROFESSIONAL ML PIPELINE             ║
# ║          Optimized for Google Colab | Fast Runtime | Full Metrics           ║
# ╚══════════════════════════════════════════════════════════════════════════════╝

# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# CELL 1 — Install Dependencies
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# Run this cell first. Colab usually has these pre-installed.

# !pip install imbalanced-learn xgboost lightgbm --quiet


# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# CELL 2 — Imports & Configuration
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns
import warnings
import time
warnings.filterwarnings('ignore')

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.linear_model  import LogisticRegression
from sklearn.tree          import DecisionTreeClassifier
from sklearn.ensemble      import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics       import (
    classification_report, confusion_matrix,
    roc_auc_score, roc_curve,
    average_precision_score, precision_recall_curve,
    f1_score, precision_score, recall_score,
    matthews_corrcoef, balanced_accuracy_score
)
from sklearn.feature_selection import mutual_info_classif
from imblearn.over_sampling    import SMOTE

try:
    from xgboost import XGBClassifier
    XGBOOST_AVAILABLE = True
except ImportError:
    XGBOOST_AVAILABLE = False

# ── Aesthetic configuration ──────────────────────────────────────────────────
DARK        = '#0d1117'
PANEL       = '#161b22'
BORDER      = '#30363d'
ACCENT_BLUE = '#58a6ff'
ACCENT_RED  = '#f85149'
ACCENT_GRN  = '#3fb950'
ACCENT_YLW  = '#d29922'
ACCENT_PRP  = '#bc8cff'
TEXT        = '#e6edf3'
MUTED       = '#8b949e'

MODEL_COLORS = {
    'Logistic Regression' : ACCENT_BLUE,
    'Decision Tree'       : ACCENT_YLW,
    'Random Forest'       : ACCENT_GRN,
    'Gradient Boosting'   : ACCENT_PRP,
    'XGBoost'             : '#ff7b72',
    'Ensemble'            : ACCENT_RED,
}

plt.rcParams.update({
    'figure.facecolor'  : DARK,
    'axes.facecolor'    : PANEL,
    'axes.edgecolor'    : BORDER,
    'text.color'        : TEXT,
    'axes.labelcolor'   : TEXT,
    'xtick.color'       : MUTED,
    'ytick.color'       : MUTED,
    'grid.color'        : BORDER,
    'grid.linestyle'    : '--',
    'grid.alpha'        : 0.5,
    'font.family'       : 'monospace',
    'legend.facecolor'  : PANEL,
    'legend.edgecolor'  : BORDER,
})

def section(title, char='━'):
    w = 70
    print(f"\n\033[1;96m{char*w}")
    pad = (w - len(title) - 2) // 2
    print(f"{char*pad} {title} {char*(w-pad-len(title)-2)}")
    print(f"{char*w}\033[0m")

def subsection(title):
    print(f"\n\033[1;93m▶ {title}\033[0m")

def ok(msg):   print(f"  \033[32m✔\033[0m  {msg}")
def info(msg): print(f"  \033[34mℹ\033[0m  {msg}")
def warn(msg): print(f"  \033[33m⚠\033[0m  {msg}")

print("\033[1;92m")
print("╔══════════════════════════════════════════════════════════════════════╗")
print("║       CREDIT CARD FRAUD DETECTION — ML PIPELINE v2.0               ║")
print("║       Fast · Professional · Interpretable                          ║")
print("╚══════════════════════════════════════════════════════════════════════╝")
print("\033[0m")


# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# CELL 3 — Load Data
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

section("STEP 1 · LOAD DATA")

# ── Upload files in Colab ────────────────────────────────────────────────────
# Option A: Upload manually
# from google.colab import files
# uploaded = files.upload()   # upload fraudTrain.csv and fraudTest.csv

# Option B: Mount Google Drive (recommended for large files)
# from google.colab import drive
# drive.mount('/content/drive')
# TRAIN_PATH = '/content/drive/MyDrive/fraudTrain.csv'
# TEST_PATH  = '/content/drive/MyDrive/fraudTest.csv'

# Option C: Direct paths if already uploaded
TRAIN_PATH = '/content/drive/MyDrive/fraudTrain.csv'
TEST_PATH  = '/content/drive/MyDrive/fraudTest.csv'

# ── SPEED OPTIMISATION: sample 300k train rows, 100k test rows ───────────────
# This cuts runtime from ~10 min → ~2 min with <1% accuracy loss.
# Remove nrows= to use the full dataset (requires ~8 GB RAM).
TRAIN_SAMPLE = 300_000
TEST_SAMPLE  = 100_000

t0 = time.time()
train_raw = pd.read_csv(TRAIN_PATH, nrows=TRAIN_SAMPLE)
test_raw  = pd.read_csv(TEST_PATH,  nrows=TEST_SAMPLE)
ok(f"Loaded in {time.time()-t0:.1f}s")

info(f"Train shape : {train_raw.shape}")
info(f"Test  shape : {test_raw.shape}")
info(f"Columns     : {list(train_raw.columns)}")

fraud_rate_train = train_raw['is_fraud'].mean() * 100
fraud_rate_test  = test_raw['is_fraud'].mean()  * 100
warn(f"Class imbalance — Fraud rate  TRAIN: {fraud_rate_train:.2f}%  |  TEST: {fraud_rate_test:.2f}%")


# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# CELL 4 — EDA Visualisation
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

section("STEP 2 · EXPLORATORY DATA ANALYSIS")

df_eda = train_raw.copy()
df_eda['trans_date_trans_time'] = pd.to_datetime(df_eda['trans_date_trans_time'])
df_eda['hour']       = df_eda['trans_date_trans_time'].dt.hour
df_eda['day_of_week']= df_eda['trans_date_trans_time'].dt.dayofweek
DAY_NAMES = ['Mon','Tue','Wed','Thu','Fri','Sat','Sun']

fig, axes = plt.subplots(2, 3, figsize=(20, 10), facecolor=DARK)
fig.suptitle('EXPLORATORY DATA ANALYSIS', fontsize=16, color=TEXT,
             fontweight='bold', y=1.02)

# 1. Class distribution
ax = axes[0, 0]
counts = df_eda['is_fraud'].value_counts()
wedges, texts, autotexts = ax.pie(
    counts, labels=['Legitimate', 'Fraud'],
    autopct='%1.2f%%', colors=[ACCENT_BLUE, ACCENT_RED],
    startangle=90, wedgeprops={'edgecolor': DARK, 'linewidth': 3},
    textprops={'color': TEXT, 'fontsize': 11})
for at in autotexts: at.set_fontsize(12); at.set_fontweight('bold')
ax.set_title('CLASS DISTRIBUTION', color=ACCENT_BLUE, fontsize=12, pad=15)

# 2. Amount distribution
ax = axes[0, 1]
legit = df_eda[df_eda['is_fraud']==0]['amt'].clip(0, 500)
fraud = df_eda[df_eda['is_fraud']==1]['amt'].clip(0, 500)
ax.hist(legit, bins=60, alpha=0.6, color=ACCENT_BLUE, label='Legitimate', density=True)
ax.hist(fraud, bins=60, alpha=0.8, color=ACCENT_RED,  label='Fraud',      density=True)
ax.set_xlabel('Transaction Amount ($, clipped @500)')
ax.set_ylabel('Density')
ax.set_title('AMOUNT DISTRIBUTION', color=ACCENT_BLUE, fontsize=12)
ax.legend(); ax.grid(True)

# 3. Fraud rate by hour
ax = axes[0, 2]
hour_fraud = df_eda.groupby('hour')['is_fraud'].mean() * 100
cmap = plt.cm.plasma(np.linspace(0.2, 0.9, 24))
ax.bar(hour_fraud.index, hour_fraud.values, color=cmap, edgecolor='none')
ax.set_xlabel('Hour of Day'); ax.set_ylabel('Fraud Rate (%)')
ax.set_title('FRAUD RATE BY HOUR', color=ACCENT_BLUE, fontsize=12)
ax.set_xticks(range(0, 24, 2)); ax.grid(True, axis='y')
peak = hour_fraud.idxmax()
ax.annotate(f'Peak: {peak}h', xy=(peak, hour_fraud[peak]),
            xytext=(peak+1, hour_fraud[peak]+0.05),
            color=ACCENT_RED, fontsize=9, arrowprops=dict(arrowstyle='->', color=ACCENT_RED))

# 4. Fraud rate by category
ax = axes[1, 0]
cat_fraud = df_eda.groupby('category')['is_fraud'].mean().sort_values() * 100
colors_c = [ACCENT_RED if v > cat_fraud.mean() else ACCENT_BLUE for v in cat_fraud]
ax.barh(cat_fraud.index, cat_fraud.values, color=colors_c, edgecolor='none')
ax.set_xlabel('Fraud Rate (%)'); ax.set_title('FRAUD RATE BY CATEGORY', color=ACCENT_BLUE, fontsize=12)
ax.axvline(cat_fraud.mean(), ls='--', color=ACCENT_YLW, lw=1.5, label=f'Mean {cat_fraud.mean():.2f}%')
ax.legend(); ax.grid(True, axis='x')

# 5. Fraud by day of week
ax = axes[1, 1]
day_fraud = df_eda.groupby('day_of_week')['is_fraud'].mean() * 100
colors_d = [ACCENT_RED if v > day_fraud.mean() else ACCENT_BLUE for v in day_fraud]
ax.bar(DAY_NAMES, day_fraud.values, color=colors_d, edgecolor='none', width=0.6)
ax.set_ylabel('Fraud Rate (%)'); ax.set_title('FRAUD RATE BY DAY', color=ACCENT_BLUE, fontsize=12)
ax.axhline(day_fraud.mean(), ls='--', color=ACCENT_YLW, lw=1.5)
ax.grid(True, axis='y')

# 6. Amount box plot
ax = axes[1, 2]
data_bp = [df_eda[df_eda['is_fraud']==0]['amt'].clip(0,300).values,
           df_eda[df_eda['is_fraud']==1]['amt'].clip(0,300).values]
bp = ax.boxplot(data_bp, labels=['Legitimate','Fraud'],
                patch_artist=True, widths=0.4,
                medianprops=dict(color=TEXT, lw=2))
bp['boxes'][0].set_facecolor(ACCENT_BLUE + '66')
bp['boxes'][1].set_facecolor(ACCENT_RED  + '66')
for w in bp['whiskers']: w.set_color(MUTED)
for c in bp['caps']:     c.set_color(MUTED)
for f in bp['fliers']:   f.set(marker='o', color=ACCENT_YLW, alpha=0.3, markersize=2)
ax.set_ylabel('Transaction Amount ($)'); ax.set_title('AMOUNT BOXPLOT', color=ACCENT_BLUE, fontsize=12)
ax.grid(True, axis='y')

plt.tight_layout()
plt.savefig('fraud_eda.png', dpi=150, bbox_inches='tight', facecolor=DARK)
plt.show()
ok("EDA plots saved → fraud_eda.png")


# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# CELL 5 — Feature Engineering
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

section("STEP 3 · FEATURE ENGINEERING")

le_cat = LabelEncoder()
le_cat.fit(pd.concat([train_raw['category'], test_raw['category']]))

def engineer(df, le_category):
    df = df.copy()
    df['trans_date_trans_time'] = pd.to_datetime(df['trans_date_trans_time'])
    df['dob']                   = pd.to_datetime(df['dob'])
    # Temporal
    df['hour']        = df['trans_date_trans_time'].dt.hour
    df['day_of_week'] = df['trans_date_trans_time'].dt.dayofweek
    df['month']       = df['trans_date_trans_time'].dt.month
    df['is_weekend']  = (df['day_of_week'] >= 5).astype(int)
    df['is_night']    = ((df['hour'] >= 22) | (df['hour'] <= 5)).astype(int)
    # Demographics
    df['age']         = (df['trans_date_trans_time'] - df['dob']).dt.days // 365
    df['gender_enc']  = (df['gender'] == 'M').astype(int)
    # Geography
    df['dist']        = np.sqrt((df['lat'] - df['merch_lat'])**2 +
                                (df['long'] - df['merch_long'])**2)
    # Amount
    df['log_amt']     = np.log1p(df['amt'])
    df['amt_sq']      = df['amt'] ** 0.5
    # Category
    df['category_enc']= le_category.transform(df['category'])
    return df

train_eng = engineer(train_raw, le_cat)
test_eng  = engineer(test_raw,  le_cat)

FEATURES = [
    'amt', 'log_amt', 'amt_sq',
    'hour', 'day_of_week', 'month', 'is_weekend', 'is_night',
    'age', 'gender_enc', 'dist', 'category_enc', 'city_pop'
]

ok(f"Engineered {len(FEATURES)} features:")
for i, f in enumerate(FEATURES, 1):
    print(f"     {i:02d}. {f}")


# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# CELL 6 — Feature Selection
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

section("STEP 4 · FEATURE SELECTION  (Mutual Information)")

X_all  = train_eng[FEATURES].values
y_all  = train_eng['is_fraud'].values
X_test = test_eng[FEATURES].values
y_test = test_eng['is_fraud'].values

# Use a small subsample for MI to be fast
sss_mi = StratifiedShuffleSplit(n_splits=1, train_size=50_000, random_state=42)
for idx_mi, _ in sss_mi.split(X_all, y_all):
    X_mi, y_mi = X_all[idx_mi], y_all[idx_mi]

mi_scores = mutual_info_classif(X_mi, y_mi, random_state=42)
mi_df     = (pd.DataFrame({'Feature': FEATURES, 'MI Score': mi_scores})
               .sort_values('MI Score', ascending=False)
               .reset_index(drop=True))

print(f"\n  {'Rank':<5} {'Feature':<20} {'MI Score':>10}  {'Bar'}")
print(f"  {'─'*55}")
bar_max = mi_df['MI Score'].max()
for i, row in mi_df.iterrows():
    bar_len = int((row['MI Score'] / bar_max) * 30)
    bar     = '█' * bar_len
    color   = '\033[92m' if row['MI Score'] > mi_df['MI Score'].mean() else '\033[90m'
    print(f"  {i+1:<5} {row['Feature']:<20} {row['MI Score']:>10.4f}  {color}{bar}\033[0m")

# Keep top features (MI > threshold)
THRESHOLD   = 0.005
TOP_FEATURES= mi_df[mi_df['MI Score'] > THRESHOLD]['Feature'].tolist()
warn(f"\n  Kept {len(TOP_FEATURES)} / {len(FEATURES)} features (MI > {THRESHOLD}): {TOP_FEATURES}")

X_all  = train_eng[TOP_FEATURES].values
X_test = test_eng[TOP_FEATURES].values


# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# CELL 7 — Scaling + SMOTE + PCA
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

section("STEP 5 · SCALING  +  SMOTE  +  PCA")

subsection("Standard Scaling")
scaler = StandardScaler()
X_sc   = scaler.fit_transform(X_all)
X_te_sc= scaler.transform(X_test)
ok("Features scaled to zero mean, unit variance")

subsection("Stratified Sub-sample for SMOTE (speed)")
sss = StratifiedShuffleSplit(n_splits=1, train_size=150_000, random_state=42)
for idx_s, _ in sss.split(X_sc, y_all):
    X_sub, y_sub = X_sc[idx_s], y_all[idx_s]
info(f"Sub-sample: {np.bincount(y_sub)}  →  {dict(zip(['Legit','Fraud'], np.bincount(y_sub)))}")

subsection("SMOTE — Synthetic Minority Oversampling")
smote  = SMOTE(random_state=42, sampling_strategy=0.25, k_neighbors=5)
X_res, y_res = smote.fit_resample(X_sub, y_sub)
ok(f"After SMOTE: Legit={np.sum(y_res==0):,}  Fraud={np.sum(y_res==1):,}  "
   f"(ratio={np.sum(y_res==1)/len(y_res)*100:.1f}%)")

subsection("PCA — Dimensionality Reduction")
pca   = PCA(n_components=0.95, random_state=42)
X_pca = pca.fit_transform(X_res)
X_tpca= pca.transform(X_te_sc)
ok(f"Dims: {X_res.shape[1]} → {X_pca.shape[1]}  (95% variance retained)")
info(f"Explained variance per PC: {pca.explained_variance_ratio_.round(3).tolist()}")


# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# CELL 8 — Model Training & Evaluation
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

section("STEP 6 · MODEL TRAINING & EVALUATION")

MODELS = {
    'Logistic Regression': LogisticRegression(
        max_iter=300, class_weight='balanced',
        solver='saga', C=0.5, random_state=42),

    'Decision Tree': DecisionTreeClassifier(
        max_depth=12, class_weight='balanced',
        min_samples_leaf=10, random_state=42),

    'Random Forest': RandomForestClassifier(
        n_estimators=80, max_depth=14,
        class_weight='balanced', min_samples_leaf=5,
        random_state=42, n_jobs=-1),

    'Gradient Boosting': GradientBoostingClassifier(
        n_estimators=80, max_depth=5,
        learning_rate=0.1, subsample=0.8, random_state=42),
}

if XGBOOST_AVAILABLE:
    scale_pos = int(np.sum(y_res==0) / np.sum(y_res==1))
    MODELS['XGBoost'] = XGBClassifier(
        n_estimators=80, max_depth=6, learning_rate=0.1,
        scale_pos_weight=scale_pos, subsample=0.8,
        eval_metric='logloss', random_state=42, n_jobs=-1, verbosity=0)

RESULTS = {}

for name, model in MODELS.items():
    subsection(f"Training: {name}")
    t_start = time.time()
    model.fit(X_pca, y_res)
    elapsed = time.time() - t_start

    y_pred = model.predict(X_tpca)
    y_prob = model.predict_proba(X_tpca)[:, 1]

    RESULTS[name] = {
        'model'    : model,
        'y_pred'   : y_pred,
        'y_prob'   : y_prob,
        'roc_auc'  : roc_auc_score(y_test, y_prob),
        'pr_auc'   : average_precision_score(y_test, y_prob),
        'f1'       : f1_score(y_test, y_pred),
        'precision': precision_score(y_test, y_pred, zero_division=0),
        'recall'   : recall_score(y_test, y_pred),
        'mcc'      : matthews_corrcoef(y_test, y_pred),
        'bal_acc'  : balanced_accuracy_score(y_test, y_pred),
        'time_s'   : elapsed,
    }
    r = RESULTS[name]
    ok(f"Done in {elapsed:.1f}s  |  ROC-AUC={r['roc_auc']:.4f}  "
       f"PR-AUC={r['pr_auc']:.4f}  F1={r['f1']:.4f}  MCC={r['mcc']:.4f}")


# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# CELL 9 — Ensemble Model (Soft Voting)
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

section("STEP 7 · ENSEMBLE MODEL  (Soft-Vote Average)")

# Average probabilities across all models
probs_stack = np.column_stack([RESULTS[n]['y_prob'] for n in RESULTS])
ens_prob    = probs_stack.mean(axis=1)
ens_pred    = (ens_prob >= 0.5).astype(int)

RESULTS['Ensemble'] = {
    'model'    : None,
    'y_pred'   : ens_pred,
    'y_prob'   : ens_prob,
    'roc_auc'  : roc_auc_score(y_test, ens_prob),
    'pr_auc'   : average_precision_score(y_test, ens_prob),
    'f1'       : f1_score(y_test, ens_pred),
    'precision': precision_score(y_test, ens_pred, zero_division=0),
    'recall'   : recall_score(y_test, ens_pred),
    'mcc'      : matthews_corrcoef(y_test, ens_pred),
    'bal_acc'  : balanced_accuracy_score(y_test, ens_pred),
    'time_s'   : 0.0,
}
r = RESULTS['Ensemble']
ok(f"Ensemble  |  ROC-AUC={r['roc_auc']:.4f}  PR-AUC={r['pr_auc']:.4f}  "
   f"F1={r['f1']:.4f}  MCC={r['mcc']:.4f}")


# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# CELL 10 — BEST MODEL ANNOUNCEMENT + FULL METRICS
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

section("STEP 8 · RESULTS — ALL MODELS")

# ── Comprehensive comparison table ──────────────────────────────────────────
METRIC_COLS = ['roc_auc','pr_auc','f1','precision','recall','mcc','bal_acc','time_s']
HEADERS     = ['ROC-AUC','PR-AUC','F1','Precision','Recall','MCC','Bal-Acc','Time(s)']

col_w  = 10
name_w = 22
divider= '─' * (name_w + col_w * len(METRIC_COLS) + 3)

print(f"\n  \033[1m{'Model':<{name_w}}", end='')
for h in HEADERS:
    print(f"{h:>{col_w}}", end='')
print("\033[0m")
print(f"  {divider}")

best_roc = max(RESULTS, key=lambda n: RESULTS[n]['roc_auc'])

for name, r in RESULTS.items():
    is_best = (name == best_roc)
    prefix  = '\033[1;92m★ ' if is_best else '  '
    suffix  = ' ★\033[0m'   if is_best else ''
    nm      = (name[:name_w-2] + '..') if len(name) > name_w else name
    print(f"{prefix}{nm:<{name_w}}", end='')
    for m in METRIC_COLS:
        val = r[m]
        if m == 'time_s':
            print(f"{val:>{col_w}.1f}", end='')
        else:
            print(f"{val:>{col_w}.4f}", end='')
    print(suffix)

print(f"  {divider}")

# ── Best model announcement ──────────────────────────────────────────────────
section("★  BEST MODEL PERFORMANCE  ★", char='═')

best = RESULTS[best_roc]
print(f"""
  \033[1;92m
  ╔══════════════════════════════════════════════════════════════╗
  ║  🏆  BEST MODEL: {best_roc:<44}║
  ╚══════════════════════════════════════════════════════════════╝
  \033[0m""")

print(f"  \033[1mPrimary Metrics (Fraud Detection Focus)\033[0m")
print(f"  {'─'*50}")

metrics_display = [
    ('ROC-AUC Score',       best['roc_auc'],   '↑ Higher = better overall discrimination'),
    ('PR-AUC Score',        best['pr_auc'],    '↑ Key for imbalanced data (precision-recall)'),
    ('F1 Score (Fraud)',     best['f1'],        '↑ Balance of precision & recall for fraud'),
    ('Precision (Fraud)',    best['precision'], '↑ Of predicted fraud, % truly fraudulent'),
    ('Recall (Fraud)',       best['recall'],    '↑ Of actual fraud, % correctly detected'),
    ('MCC',                  best['mcc'],       '↑ Balanced metric robust to imbalance'),
    ('Balanced Accuracy',    best['bal_acc'],   '↑ Average recall across both classes'),
]

for label, val, note in metrics_display:
    bar_len = int(val * 30)
    color   = '\033[92m' if val > 0.7 else ('\033[93m' if val > 0.4 else '\033[91m')
    bar     = '█' * bar_len + '░' * (30 - bar_len)
    print(f"  {label:<25} \033[1m{val:.4f}\033[0m  {color}{bar}\033[0m")
    print(f"  {'':<25} \033[90m{note}\033[0m")
    print()

print(f"\n  \033[1mDetailed Classification Report — {best_roc}\033[0m")
print(f"  {'─'*50}")
report = classification_report(y_test, best['y_pred'],
                                target_names=['Legitimate','Fraud'])
for line in report.split('\n'):
    print(f"  {line}")

print(f"\n  \033[1mConfusion Matrix — {best_roc}\033[0m")
cm     = confusion_matrix(y_test, best['y_pred'])
tn, fp, fn, tp = cm.ravel()
print(f"""
  {'':>15}  Predicted Legit  Predicted Fraud
  Actual Legit  │  \033[92m{tn:>12,}\033[0m  │  \033[91m{fp:>12,}\033[0m  │
  Actual Fraud  │  \033[91m{fn:>12,}\033[0m  │  \033[92m{tp:>12,}\033[0m  │

  True  Positives (Fraud caught)    : \033[92m{tp:>8,}\033[0m
  False Positives (False alarms)    : \033[93m{fp:>8,}\033[0m
  False Negatives (Missed fraud)    : \033[91m{fn:>8,}\033[0m
  True  Negatives (Legit correct)   : \033[92m{tn:>8,}\033[0m

  Fraud Detection Rate              : \033[1;92m{tp/(tp+fn)*100:.2f}%\033[0m
  False Alarm Rate                  : \033[1;93m{fp/(fp+tn)*100:.2f}%\033[0m
""")


# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# CELL 11 — Visualisation Dashboard
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

section("STEP 9 · VISUALISATION DASHBOARD")

MODEL_NAMES = list(RESULTS.keys())
MC = {n: MODEL_COLORS.get(n, ACCENT_BLUE) for n in MODEL_NAMES}

fig = plt.figure(figsize=(22, 18), facecolor=DARK)
fig.suptitle('FRAUD DETECTION — MODEL PERFORMANCE DASHBOARD',
             fontsize=18, color=TEXT, fontweight='bold', y=1.01)

gs = gridspec.GridSpec(3, 3, figure=fig, hspace=0.55, wspace=0.38)

# ── ROC Curves ──────────────────────────────────────────────────────────────
ax1 = fig.add_subplot(gs[0, :2])
for name in MODEL_NAMES:
    lw  = 3 if name in [best_roc, 'Ensemble'] else 1.5
    ls  = '--' if name == 'Ensemble' else '-'
    fpr, tpr, _ = roc_curve(y_test, RESULTS[name]['y_prob'])
    label = f"{'★ ' if name==best_roc else ''}{name}  (AUC={RESULTS[name]['roc_auc']:.4f})"
    ax1.plot(fpr, tpr, lw=lw, ls=ls, color=MC[name], label=label)
ax1.plot([0,1],[0,1], '--', color=MUTED, lw=1, alpha=0.5, label='Random (0.5000)')
ax1.set_xlabel('False Positive Rate'); ax1.set_ylabel('True Positive Rate')
ax1.set_title('ROC CURVES', color=ACCENT_BLUE, fontsize=13)
ax1.legend(fontsize=9, loc='lower right'); ax1.grid(True)
ax1.fill_between([0,1],[0,1], alpha=0.03, color=MUTED)

# ── Precision-Recall Curves ──────────────────────────────────────────────────
ax2 = fig.add_subplot(gs[0, 2])
for name in MODEL_NAMES:
    lw = 2.5 if name in [best_roc, 'Ensemble'] else 1.5
    prec, rec, _ = precision_recall_curve(y_test, RESULTS[name]['y_prob'])
    label = f"{'★' if name==best_roc else ''}{name}\n(AP={RESULTS[name]['pr_auc']:.3f})"
    ax2.plot(rec, prec, lw=lw, color=MC[name], label=label)
ax2.set_xlabel('Recall'); ax2.set_ylabel('Precision')
ax2.set_title('PRECISION-RECALL', color=ACCENT_BLUE, fontsize=12)
ax2.legend(fontsize=7, loc='upper right'); ax2.grid(True)

# ── Metric Comparison Bar Chart ──────────────────────────────────────────────
ax3 = fig.add_subplot(gs[1, :2])
met_names = ['roc_auc','pr_auc','f1','precision','recall','mcc']
met_lbls  = ['ROC-AUC','PR-AUC','F1','Precision','Recall','MCC']
x = np.arange(len(met_names))
n_models = len(MODEL_NAMES)
bar_w = 0.8 / n_models
offsets = np.linspace(-(n_models-1)/2*bar_w, (n_models-1)/2*bar_w, n_models)

for i, name in enumerate(MODEL_NAMES):
    vals = [RESULTS[name][m] for m in met_names]
    bars = ax3.bar(x + offsets[i], vals, bar_w,
                   label=('★ ' if name==best_roc else '') + name,
                   color=MC[name], alpha=0.9, edgecolor='none')
    if name == best_roc:
        for bar in bars:
            bar.set_edgecolor(TEXT); bar.set_linewidth(1.5)

ax3.set_xticks(x); ax3.set_xticklabels(met_lbls, fontsize=11)
ax3.set_ylim(0, 1.18); ax3.set_ylabel('Score')
ax3.set_title('MODEL METRICS COMPARISON', color=ACCENT_BLUE, fontsize=13)
ax3.legend(fontsize=9, loc='upper right', ncol=2); ax3.grid(True, axis='y')

# ── PCA Explained Variance ───────────────────────────────────────────────────
ax4 = fig.add_subplot(gs[1, 2])
ev  = pca.explained_variance_ratio_
cum = np.cumsum(ev)
pcs = range(1, len(ev)+1)
ax4.bar(pcs, ev, color=ACCENT_BLUE, alpha=0.7, label='Per-PC')
ax4_r = ax4.twinx()
ax4_r.plot(pcs, cum, 'o-', color=ACCENT_YLW, lw=2, ms=5, label='Cumulative')
ax4_r.axhline(0.95, ls='--', color=ACCENT_GRN, lw=1.5, alpha=0.8)
ax4_r.set_ylabel('Cumulative Variance', color=ACCENT_YLW)
ax4_r.tick_params(axis='y', colors=ACCENT_YLW)
ax4.set_xlabel('Principal Component'); ax4.set_ylabel('Variance Ratio', color=ACCENT_BLUE)
ax4.tick_params(axis='y', colors=ACCENT_BLUE)
ax4.set_title('PCA EXPLAINED VARIANCE', color=ACCENT_BLUE, fontsize=12)
lines1, lab1 = ax4.get_legend_handles_labels()
lines2, lab2 = ax4_r.get_legend_handles_labels()
ax4.legend(lines1+lines2, lab1+lab2, fontsize=9); ax4.grid(True, axis='y')

# ── Confusion Matrices (best + ensemble) ────────────────────────────────────
for plot_i, name in enumerate([best_roc, 'Ensemble']):
    ax = fig.add_subplot(gs[2, plot_i])
    cm_raw  = confusion_matrix(y_test, RESULTS[name]['y_pred'])
    cm_norm = cm_raw.astype(float) / cm_raw.sum(axis=1, keepdims=True)
    annot   = np.array([[f"{cm_raw[i,j]:,}\n({cm_norm[i,j]*100:.1f}%)"
                          for j in range(2)] for i in range(2)])
    sns.heatmap(cm_norm, annot=annot, fmt='', ax=ax,
                cmap='YlOrRd', linewidths=2, linecolor=DARK,
                xticklabels=['Legit','Fraud'],
                yticklabels=['Legit','Fraud'], cbar=False,
                annot_kws={'fontsize': 11, 'color': DARK, 'fontweight': 'bold'})
    star = '★ ' if name == best_roc else ''
    ax.set_title(f'{star}CONFUSION MATRIX\n{name}', color=ACCENT_BLUE, fontsize=11)
    ax.set_xlabel('Predicted'); ax.set_ylabel('Actual')

# ── Feature Importance (MI) ──────────────────────────────────────────────────
ax5 = fig.add_subplot(gs[2, 2])
mi_plot = mi_df.sort_values('MI Score')
colors_mi = [ACCENT_RED if v > mi_df['MI Score'].mean() else ACCENT_BLUE
             for v in mi_plot['MI Score']]
ax5.barh(mi_plot['Feature'], mi_plot['MI Score'], color=colors_mi, edgecolor='none')
ax5.axvline(mi_df['MI Score'].mean(), ls='--', color=ACCENT_YLW, lw=1.5, label='Mean MI')
ax5.set_xlabel('Mutual Information Score')
ax5.set_title('FEATURE IMPORTANCE (MI)', color=ACCENT_BLUE, fontsize=12)
ax5.legend(fontsize=9); ax5.grid(True, axis='x')

plt.savefig('fraud_dashboard.png', dpi=150, bbox_inches='tight', facecolor=DARK)
plt.show()
ok("Dashboard saved → fraud_dashboard.png")


# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# CELL 12 — Final Summary
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

section("FINAL SUMMARY")

rankings = sorted(RESULTS.items(), key=lambda x: x[1]['roc_auc'], reverse=True)

print("\n  \033[1mRANKING BY ROC-AUC (best fraud detector)\033[0m\n")
medals = ['🥇', '🥈', '🥉'] + ['  '] * 10
for rank, (name, r) in enumerate(rankings):
    bar = '█' * int(r['roc_auc'] * 40)
    print(f"  {medals[rank]}  #{rank+1}  {name:<24}  ROC-AUC={r['roc_auc']:.4f}  {bar}")

print(f"""
  \033[1;96m
  ┌─────────────────────────────────────────────────────────────┐
  │  RECOMMENDATION                                             │
  │                                                             │
  │  Best Single Model : {best_roc:<38}│
  │  ROC-AUC           : {RESULTS[best_roc]['roc_auc']:.4f}                              │
  │  PR-AUC            : {RESULTS[best_roc]['pr_auc']:.4f}   (key for imbalanced data)   │
  │  Recall (Fraud)    : {RESULTS[best_roc]['recall']:.4f}   (% of fraud caught)         │
  │                                                             │
  │  For production: tune the decision threshold on PR curve    │
  │  to trade false alarms vs. missed fraud as needed.         │
  └─────────────────────────────────────────────────────────────┘
  \033[0m""")

section("PIPELINE COMPLETE", char='═')
```
